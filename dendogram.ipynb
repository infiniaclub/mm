```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

x = [8, 5, 10, 7, 3, 11, 14 , 16, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 12, 21, 21]

data = list(zip(x, y))
model = linkage(data, method="ward", metric="euclidean")
dendrogram(model)
plt.show()
```
