```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import confusion_matrix, classification_report

df=pd.read_csv('https://gist.githubusercontent.com/DiogoRibeiro7/c6590d0cf119e87c39e31c21a9c0f3a8/raw/4a8e3da267a0c1f0d650901d8295a5153bde8b21/PlayTennis.csv')
df.head()
le = LabelEncoder()
for i in df.columns:
    df[i] = le.fit_transform(df[i])
print(df.head())
x = df.drop("Play Tennis", axis=1)
y= df["Play Tennis"]
model=DecisionTreeClassifier()
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
accuracy = model.score(X_test,y_test)
print(f"accuracy is {accuracy*100}%")
plot_tree(model,filled=True,feature_names=df.columns,class_names=le.classes_)
plt.show()
model.predict([[1,2,0,1]])
print(f"confusion matrix : \n {confusion_matrix(y_test,y_pred)}")
print(f"confusion matrix : \n {classification_report(y_test,y_pred)}")
```
