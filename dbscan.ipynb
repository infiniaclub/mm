```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import seaborn as sns


x = np.array([3, 4, 5, 7, 8, 5, 8, 9, 3, 3, 2, 5])
y = np.array([7, 6, 5, 5, 4, 1, 3, 3, 7, 5, 4, 9])

data = np.column_stack((x, y))

model = DBSCAN(eps=1.9, min_samples=4)
model.fit(data)
# Get coordinates of core points
core_points = data[model.core_sample_indices_]

# Get coordinates of border points
border_points = data[model.labels_ != -1]  # Exclude noise points
# Get coordinates of noise points
noise_points = data[model.labels_ == -1]
# Print core point coordinates
print(f"Core Point Coordinates: \n {core_points}")
print(f"Border Point Coordinates: \n {border_points}")
print(f"Noise Point Coordinates: \n {noise_points}")

# Create a Seaborn scatter plot with cluster labels and noise points labeled as -1
sns.scatterplot(x=x, y=y, hue=model.labels_, palette="coolwarm", s=100)
```
